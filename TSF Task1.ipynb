#Data: http://bit.ly/w-data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
print('Libraries imported successfully')
```

```python
url = "http://bit.ly/w-data"
ab=pd.read_csv(url)
print("Data imported successfully")
```

```python
ab.plot(kind='scatter', x='Hours', y='Scores', color='r')
plt.title('Hours Vs Percentage')
plt.xlabel('Hours studied')
plt.ylabel('% Score')
plt.show()
```

```python
x=np.asanyarray(ab[['Hours']])
y=np.asanyarray(ab['Scores'])
regressor = LinearRegression()
regressor.fit(x,y)
ab.plot(kind='scatter', x='Hours', y='Scores', color='r')
plt.plot(x, regressor.coef_[0]*x + regressor.intercept_, color='b')
plt.title('Hours Vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('% Score')
plt.grid()
plt.show()
```

```python
hrs=9.25
p_score=regressor.predict([[hrs]])
print(f'No of hours = {hrs}')
print(f'predicted score = {p_score[0]}')
```
